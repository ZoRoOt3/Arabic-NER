```python
def getNerTag(dataitems):
    dic={}
    count=-1;
    for data in dataitems:
        count=count+1
        for sentence in data['paragraphs'][0]['sentences']:
            tokens=sentence['tokens']
            for token in tokens:
                nertag=token['ner']
                #if nertag not in dic:
                if('B-FAC' in nertag):
                    #print(count)
                    token['ner']='B-LOC'                    
                elif('I-FAC' in nertag):
                    token['ner']='I-LOC'
                elif('L-FAC' in nertag):
                    token['ner']='L-LOC'
                elif('U-FAC' in nertag):
                    token['ner']='U-LOC'
                elif('B-GPE' in nertag):
                    token['ner']='B-LOC'
                elif('I-GPE' in nertag):
                    token['ner']='I-LOC'
                elif('L-GPE' in nertag):
                    token['ner']='L-LOC'
                elif('U-GPE' in nertag):
                    token['ner']='U-LOC'
                elif('B-LOC' in nertag):
                    token['ner']='B-LOC'
                elif('I-LOC' in nertag):
                    token['ner']='I-LOC'
                elif('L-LOC' in nertag):
                    token['ner']='L-LOC'
                elif('U-LOC' in nertag):
                    token['ner']='U-LOC'
                elif('B-PERS' in nertag):
                    token['ner']='B-PERSON'
                elif('I-PERS' in nertag):
                    token['ner']='I-PERSON'
                elif('L-PERS' in nertag):
                    token['ner']='L-PERSON'
                elif('U-PERS' in nertag):
                    token['ner']='U-PERSON'
                elif('B-ORG' in nertag):
                    token['ner']='B-ORG'
                elif('I-ORG' in nertag):
                    token['ner']='I-ORG'
                elif('L-ORG' in nertag):
                    token['ner']='L-ORG'
                elif('U-ORG' in nertag):
                    token['ner']='U-ORG'
                elif('U-MISC' in nertag):
                    token['ner']='U-MISC'
                elif('B-MISC' in nertag):
                    token['ner']='B-MISC'
                elif('I-MISC' in nertag):
                    token['ner']='I-MISC'
                elif('L-MISC' in nertag):
                    token['ner']='L-MISC'
                elif('U-ts' in nertag):
                    token['ner']='O'
                elif(nertag=='O'):
                    token['ner']='O'
                elif(nertag=='-'):
                    token['ner']='O'
                elif("B-" in nertag):
                    token['ner']='B-MISC'
                elif("I-" in nertag):
                    token['ner']='I-MISC'
                elif("L-" in nertag):
                    token['ner']="L-MISC"
                elif("U-" in nertag):
                    token['ner']='U-MISC'
                if(token['ner'] not in dic):
                    dic[token['ner']]=1
                else:
                    dic[token['ner']]=dic[token['ner']]+1

    return dic
                
    
```

```python
def cleantext(text):
    no_diacritics = araby.strip_tashkeel(text)
    just_arabic_text = re.sub(r'\W+', '',no_diacritics ) # removes everything else but the word. This assumes there's only one word in orth 
    return just_arabic_text
    
    
```

```python
d=data4[0]['paragraphs'][0]['sentences'][0]['tokens'][0]['orth']
```

```python
def cleanupToken(data):
    token_count=0;
    for d in data:
        para=d['paragraphs'][0];
        for sen in para['sentences']:
            for tok in sen['tokens']:
                token_count+=1;
                tok['orth']=cleantext(tok['orth'])
    return token_count;
        
```
