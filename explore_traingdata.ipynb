```python
def getNerTag(dataitems):
    dic={}
    count=-1;
    for data in dataitems:
        count=count+1
        for sentence in data['paragraphs'][0]['sentences']:
            tokens=sentence['tokens']
            for token in tokens:
                nertag=token['ner']
                if nertag not in dic:
                    #if(nertag== 'B-DATE" E_OFF="5'):
                        #print(count)
                    dic[nertag]=1
                else:
                    dic[nertag]=dic[nertag]+1;
    return dic
                
    
```
